```python
COLORS_SRC_FILENAME = os.path.join("data", "colors", "filteredCorpus.csv")

corpus = ColorsCorpusReader(
    COLORS_SRC_FILENAME,
    word_count=None,
    normalize_colors=True)
examples=list(corpus.read())
```

```python
txts=[]
for i in examples:
    txt=i.parse_turns()
    txt=" , ".join(txt)
    print(txt)
    txts.append(txt)
```

```python
import spacy
import en_core_web_lg

nlp = en_core_web_lg.load()

# nlp = spacy.load("en_core_web_sm")
# doc = nlp("oopsies. I think they're both a bit neon , the greyish one that's not the browner one")

def getPOSTags(txt):
    output=[]
    doc = nlp(txt)
    for token in doc:
        output.append([token.text, token.pos_])
    return output
```

```python
df=[]
for i in tqdm(txts):
    df.extend(getPOSTags(i))

df=pd.DataFrame(df,columns=['tok','tag'])
```

```python
def getNER(txt):
    output=[]
    doc = nlp(txt)
    for token in doc.ents:
        output.append([token.text, token.label_])
    return output
```

```python
nerdf=[]
for i in tqdm(txts):
    nerdf.extend(getNER(i))
    
nerdf=pd.DataFrame(nerdf,columns=['tok','tag'])
```
